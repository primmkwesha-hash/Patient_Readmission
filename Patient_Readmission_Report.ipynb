Observation

The gender column contains two unique categories and male being the most frequent. Elective being the most common admission type and has No readmission records.

```python
# Count the number of patients by gender
df['Gender'].value_counts()
```

Observation

More male patients than female patients

```python
# Checking the age buckets
conditions = [
    df['Age'] < 25,
    df['Age'] .between (25, 40),
    df['Age'] .between (41, 60),
    df['Age'] .between (61, 80),
    df['Age'] > 80

]
choices = [
    'Youth',
    'Young Adults',
    'Adults',
    'Senior',
    'Old'

]
df['Age Group'] = np.select(conditions ,choices, default='Unknown')
```

```python
# count patients per age group
df['Age Group'].value_counts()
```

Observation

Seniors make up the largest age group in the dataset 835 patients, followed by Adults 824 patients.

```python
# Checking the age groups by Admission Type (Emergency vs Elective)
pd.crosstab(df['Admission Type'], df['Age Group'])
```

Observation

There are fewer emergency patients in each age group than elective patients.
